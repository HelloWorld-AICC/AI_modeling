## 지식인

```python
import os, json
file_path = './jisikin.json'

# 기존 파일 읽기 (파일이 없는 경우 새로 생성)
if os.path.exists(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)  # 기존 데이터를 읽어옴
        except json.JSONDecodeError:
            data = []  # 파일이 비어있으면 빈 리스트로 초기화
else:
    data = []
```

```python
new_data = []

for row in data:
    new_dict = {}
    new_dict["title"] = row['title'] + "\n" + row['contents']
    new_dict["contents"] = row['answers']
    new_dict["date"] = row['date']
    new_dict["url"] = "https://kin.naver.com/userinfo/answerList.naver?isSearch=true&isWorry=false&u=LozJsi3cGuRd5Y5PklLl8GwBQI1V%2BpaEnpueNikIQss%3D&section=qna&sd=answer&query=%EC%99%B8%EA%B5%AD%EC%9D%B8&x=0&y=0"
    
    new_data.append(new_dict)
```

```python
new_data
```

```python
# 업데이트된 데이터를 다시 파일에 저장
with open('./jisikin.json', 'w', encoding='utf-8') as f:
    json.dump(new_data, f, ensure_ascii=False, indent=4)
```

## 대한법률구조공단

```python
import os, json
file_path = './대한법률구조공단_상담사례.json'

# 기존 파일 읽기 (파일이 없는 경우 새로 생성)
if os.path.exists(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)  # 기존 데이터를 읽어옴
        except json.JSONDecodeError:
            data = []  # 파일이 비어있으면 빈 리스트로 초기화
else:
    data = []
```

```python
data
```

```python
data = [row for row in data if "외국인" in row["title"]]
data
```

```python
new_data = []

for row in data:
    new_dict = {}
    new_dict["topic"] = row["main_category"] + row["sub_category"]
    new_dict["title"] = row["title"]
    new_dict["contents"] = row["question"] + "\n" + row["answer"]
    new_dict["url"] = "https://www.klac.or.kr/"
    
    new_data.append(new_dict)
```

```python
# 업데이트된 데이터를 다시 파일에 저장
with open('./klac.json', 'w', encoding='utf-8') as f:
    json.dump(new_data, f, ensure_ascii=False, indent=4)
```
